```python
import cv2
#import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('./Nature.jpeg')
```

### calcHist() and hist()
plt.calcHist([image], [channel], mask, [hist_size], range)  
plt.hist(image_variable.ravel(), bins value, range)  

Note:
- channel value is [0] for gray scale images and [0], [1] and [2] for BGR components respectively of colored image
- hist_size is bin value
- range is intensity value range thats [0, 256] for all intensity values

```python
hist=cv2.calcHist([img],[0],None,[256],[0,256])
plt.hist(img.ravel(),256,[0,256],histtype='step')
# Here, 256 value i.e. 2nd parameter is bins value 
plt.title('Image Histogram',fontweight="bold")
plt.xlabel("Intensity values")
plt.ylabel("Number of pixels")
plt.show()
```

```python
hist=cv2.calcHist([img],[0],None,[256],[0,256])
plt.hist(img.ravel(),256,[0,256],color='b')
plt.title('Image Histogram',fontweight="bold")
plt.xlabel("Intensity values")
plt.ylabel("Number of pixels")

plt.show()
```

```python
color=('b', 'g', 'r')

for i, col in enumerate(color):
    hist= cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)

plt.title('Image Histogram',fontweight="bold")
plt.xlabel("Intensity values")
plt.ylabel("Number of pixels")
plt.show()
```

```python
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hist=cv2.calcHist([gray_img],[0],None,[256],[0,256])
plt.hist(gray_img.ravel(),256,[0,256],histtype='step', color="Black")
plt.hist(img.ravel(),256,[0,256],histtype='step', color="Blue")

plt.title('Image Histogram',fontweight="bold")
plt.xlabel("Intensity values")
plt.ylabel("Number of pixels")
plt.show()
```
